```python
import pytrends
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
from pytrends.request import TrendReq
pytrend = TrendReq()
pytrend.build_payload(kw_list=['Bitcoin'])
bitcoin_geo_trend = pytrend.interest_by_region(resolution = 'COUNTRY_NAME') # resolution = 'COUNTRY_NAME' não muda nada
bitcoin_geo_trend.head()
```

```python
#obs o valor é uma escala do google
print("Venezuela : ",bitcoin_geo_trend.loc["Venezuela",])
print("Brazil : ",bitcoin_geo_trend.loc["Brazil",])
print("Argentina : ",bitcoin_geo_trend.loc["Argentina",])
print("Japan : ",bitcoin_geo_trend.loc["Japan",])
print("France : ",bitcoin_geo_trend.loc["France",])
print("Canada : ",bitcoin_geo_trend.loc["Canada",])
print("Monaco : ",bitcoin_geo_trend.loc["Monaco",])
print("Mexico : ",bitcoin_geo_trend.loc["Mexico",])
```

```python
bitcoin_geo_trend.reset_index().plot(x="geoName", y="Bitcoin", figsize=(120, 10), kind ="bar")
```
